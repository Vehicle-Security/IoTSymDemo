1. 单一目标求解
目标：求一个数字 x，满足 x + 5 = 15

```python
from z3 import *

x = Int("x")

# 2. 创建求解器
s = Solver()

# 3. 添加约束条件：x + 5 = 15
s.add(x + 5 == 15)

# 4. 检查是否有解
print("是否有解:", s.check())

# 5. 打印解
print("x =", s.model()[x])
```

2. 多目标求解
目标：求 x, y 同时满足：
x > 0
y < 5
x + y = 7

```python
from z3 import *

# 定义两个变量
x = Int("x")
y = Int("y")

s = Solver()
# 加多个约束
s.add(x > 0)
s.add(y < 5)
s.add(x + y == 7)

print(s.check())
print(s.model())
```

3. buffer 约束检查

```python
from z3 import *
buf = [BitVec(f"buf_{i}", 8) for i in range(16)]

s = Solver()
# 添加约束：buf[0] + buf[1] == 10
s.add(buf[0] + buf[1] == 10)
s.add(buf[0] > 0)
s.add(buf[1] > 0)

print(s.check())
model = s.model()
print("buf[0] =", model[buf[0]])
print("buf[1] =", model[buf[1]])
```

4. CO3例子中的检查

```python
from z3 import *
addr = BitVec("addr", 8)
length = BitVec("length", 8)

s = Solver()
s.add(addr == 0x5A)       # 地址约束
s.add(And(length > 0, length <= 4)) # 长度约束

print(s.check())
print(s.model())
```
